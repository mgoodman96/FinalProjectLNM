```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# List the variables to include based on VIF results
variables_to_include = [
    'Some HS',  # VIF < 5
    'Family Poverty',  # VIF just above 5, but you might choose to include it
    'Drive Alone',
    'Private Health Ins'
]

df['Life Expectancy_abvavg'] = pd.to_numeric(df['Life Expectancy_abvavg'], errors='coerce')
df.dropna(subset=['Life Expectancy_abvavg'], inplace=True)  
y = df['Life Expectancy_abvavg'].astype(int)  

# Select the features based on the variables to include
X = df[variables_to_include]

# Add a constant to the model (intercept term)
X = sm.add_constant(X)

# Fit the logistic regression model
model = sm.Logit(y, X).fit()

# Print the model summary, which includes the log-likelihood
print(model.summary())
```

```python
log_likelihood = model.llf

print(f'Log-Likelihood: {log_likelihood}')
```
